```python
import pandas as pd
import matplotlib.pyplot as plt

name = input("Enter student name: ")

subjects = ["Subject 1", "Subject 2", "Subject 3", "Subject 4", "Subject 5"]
marks = []

for sub in subjects:
    m = float(input(f"Enter marks for {sub}: "))
    marks.append(m)

data = pd.DataFrame({
    "Subject": subjects,
    "Marks": marks
})

total = data["Marks"].sum()
average = data["Marks"].mean()

if average >= 90:
    grade = 'A+'
elif average >= 80:
    grade = 'A'
elif average >= 70:
    grade = 'B'
elif average >= 60:
    grade = 'C'
elif average >= 50:
    grade = 'D'
else:
    grade = 'F'

print("\n----- Student Result -----")
print("Student Name:", name)
print("Total Marks:", total)
print("Average Marks:", average)
print("Grade:", grade)

plt.bar(data["Subject"], data["Marks"])
plt.title(f"{name}'s Marks Analysis")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
